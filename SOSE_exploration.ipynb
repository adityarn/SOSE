```python
def find_spatial_indices(lat=None, lon=None):
    if((lat==None) or (lon==None)):
        print("Error!! Must give both lat and lon")
    else:
        lat_ind = np.argmin(np.abs(grid['YC'][0, :] - lat))
        lon_ind = np.argmin(np.abs(grid['XC'][:, 0] - 180 - lon))
        return lon_ind, lat_ind
```
